```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys

# Attendez que la page se charge complètement (vous pouvez ajuster le temps d'attente en fonction de la vitesse de votre connexion)
time.sleep(5)

# Faites défiler la page vers le bas plusieurs fois pour charger les commentaires
for _ in range(200):  
    body = driver.find_element("tag name", "body")
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2) 

time.sleep(2)  # Attendez un court moment pour que les commentaires se chargent

# Récupérez les éléments HTML des commentaires
comment_elements = driver.find_elements(by = 'xpath', value ="//yt-formatted-string[@class='style-scope ytd-comment-renderer']")

# Affichez les commentaires
for comment_element in comment_elements:
    print(comment_element.text)

# Fermez le navigateur
#driver.quit()
```

```python
# Affichez les commentaires
l = []
for comment_element in comment_elements:
    l.append(comment_element.text)

print(len(l))
```

```python
j = []
for i in range(len(l)) :
    j.append("Grand Theft Auto VI")

print(len(j))
```

```python
import pandas as pd

df = pd.DataFrame(data = zip(j,l), columns=['Jeu','Commentaire'])

display(df)
```

```python
from langdetect import detect
import pandas as pd

# Supposons que vous avez un DataFrame 'df' avec une colonne 'comment' contenant les commentaires
# Exemple :
# df = pd.DataFrame({'comment': ['This is an English comment', 'Ceci est un commentaire en français', 'Este es un comentario en español']})

# Ajouter une colonne 'language' au DataFrame pour stocker la langue de chaque commentaire
df['Commentaire_English'] = df['Commentaire'].apply(lambda x: detect(x) if pd.notnull(x) else None)

# Filtrer les commentaires en langue anglaise
df_english_comments = df[df['Commentaire_English'] == 'en']

# Afficher les commentaires en langue anglaise
display(df_english_comments)
```
